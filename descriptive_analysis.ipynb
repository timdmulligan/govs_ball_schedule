```python
import pandas as pd
import sqlite3
```

```python
## Bring in Data
con = sqlite3.connect("pitchfork-data.db")
reviews = pd.read_sql_query("SELECT * from review_text", con)
con.close()
```

```python
reviews["score"] = reviews["score"].astype(float)
```

```python
auth_describe = reviews.groupby(["author"])["score"].describe().reset_index()
```

```python
top_auths = auth_describe[auth_describe["count"] > 50]
```

```python
top_auths.sort_values("mean")
```
